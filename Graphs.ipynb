```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
sample=pd.read_csv("music_samplesubmission.csv")
train=pd.read_csv("music_train.csv")
train=train.fillna(train.mean())
train["topic"]=train["topic"].astype(int)

X=train[['release_date', 'dating', 'violence', 'world/life', 'night/time',
       'shake the audience', 'family/gospel', 'romantic', 'communication',
       'obscene', 'music', 'movement/places', 'light/visual perceptions',
       'family/spiritual', 'like/girls', 'sadness', 'feelings', 'danceability',
       'loudness', 'acousticness', 'instrumentalness', 'valence', 'energy',
       'age',"topic"]]
Y=train["genre"]
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,random_state=0)
```

```python
print(np.shape(X_test))
```

```python
arr=[]
for i in range(50,1000,50) :            
             RForest_model = RandomForestClassifier(n_estimators=i)
             RForest_model.fit(X_train, Y_train)
             arr.append(RForest_model.score(X_test,Y_test))
```
